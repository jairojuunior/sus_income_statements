```python
for file in os.listdir("raw_files"):
    #File year identification
    year=""
    fp = io.open("raw_files/"+file, mode="r", encoding="cp1258")
    for i, line in enumerate(fp):
        if i==2:
            year = line[8:-1]
            break
    fp.close()
    print("Year: "+year)

    #Import
    df = pd.read_csv("raw_files/"+file, 
                        sep=";", decimal=".", skiprows=3, 
                        quotechar='"', encoding="cp1258")

    #Separate the city IBGE code
    df['cod_municipio'] = df['Munic-BR'].apply(lambda x: x[0:6])

    #Convert to numeric strings with different decimal char
    for col in df.columns:
        if col not in ["Munic-BR", "cod_municipio"]:
            if df[col].dtype == "object":
                try:
                    df[col] = pd.to_numeric(df[col])
                except:
                    df[col] = pd.to_numeric(df[col].apply(lambda x: x.replace(",",".")), errors="coerce")
                    
    #Operation extract
    print(df.info())
                
    #Output result
    df.to_csv("output_income_balances/health_income_balance_"+year+".csv", 
              index=False, sep=";", encoding="utf-8")
```
